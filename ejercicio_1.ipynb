```python
df=pd.read_csv('https://drive.google.com/uc?export=download&id=1r0xTsIDMpK8NUsXXkE2SlNcTIa-dF-z0')
```

```python
df=df[['puma_id','puma_name','occupation_id','occupation_name','men_20','men_30','women_20','women_30']]
df['a_20']=df['men_20']+df['women_20']
df['a_30']=df['men_30']+df['women_30']
df.head()
```

```python
#POR PUMA
df1=df.melt(id_vars=['puma_id','occupation_id'],value_vars=['men_20','men_30','women_20','women_30'])

df1['sex_id'] = df1['variable'].apply(lambda x: 'men' if x == 'men_20' or x=='men_30' else 'woman')
df1['age_id'] = df1['variable'].apply(lambda x: '0-20' if x == 'men_20' or x=='women_20' else '20-29')
df1=df1[['puma_id','occupation_id','sex_id','age_id','value']]
df1.columns=['puma_id','occupation_id','sex_id','age_id','total']
df1['sex_id']=df1['sex_id'].replace({'men' : 1,'women':2})
df1['age_id']=df1['age_id'].replace({'0-20' : 1,'20-29':2})
df1.head()
```
